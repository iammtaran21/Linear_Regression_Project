```python
#Dummies for categorical 
Bikes_categorical2=Bikes_df
for m in Bikes_categorical:
  dummies =pd.get_dummies(Bikes_df[m],drop_first=True)
  Bikes_categorical2 = pd.concat([Bikes_categorical2,dummies],axis='columns')
```

```python
#outlier analysis:
sns.boxplot(Bikes_categorical2['Rented Bike Count'])
plt.title('Distribution of Target Variable (Rented Bike Count)')
plt.show()
```

```python
#remove outlier
Q1 = np.percentile(Bikes_categorical2['Rented Bike Count'], 25, interpolation = 'midpoint')  
Q2 = np.percentile(Bikes_categorical2['Rented Bike Count'], 50, interpolation = 'midpoint')  
Q3 = np.percentile(Bikes_categorical2['Rented Bike Count'], 75, interpolation = 'midpoint')     
IQR = Q3 - Q1 
low_lim = Q1 - 1.5 * IQR 
up_lim = Q3 + 1.5 * IQR
print('low limit:', low_lim) 
print('upper limit:', up_lim) 
```

```python
Bikes_categorical2.rename(columns = {'Rented Bike Count':'RBC'}, inplace = True)
```

```python
Bikes_df2 = Bikes_categorical2[Bikes_categorical2.RBC< up_lim]
Bikes_df2=Bikes_df2[Bikes_df2.RBC> low_lim]
```

```python
from sklearn.preprocessing import StandardScaler
# Initialise the Scaler
scaler = StandardScaler()
  
# To scale data
scaler.fit(Bikes_categorical2)
#from sklearn.preprocessing import normalize
#normalize(df1)
```

```python
from sklearn.model_selection import train_test_split
train, val = train_test_split(Bikes_categorical2, test_size=0.25)
```

```python
tx=train.drop(['RBC'],axis=1)
ty=train['RBC']

vx=val.drop(['RBC'],axis=1)
vy=val['RBC']
```

```python
print(ty.shape)
print(tx.shape)
print(vy.shape)
print(vx.shape)
test=Bikes_df[6570:]
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(tx,ty)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
df_pred=lr.predict(vx)
```

```python

print("The model's metrics values are: ")
print("Accuracy: ", lr.score(vx,vy))
print("Mean Absolute Error: ",mean_absolute_error(df_pred , vy))
print("Mean Squared Error: " ,mean_squared_error(df_pred,vy))
print("R-squared score: ", r2_score(vy,df_pred))
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge
ridge.fit(tx,ty)
rp=ridge.predict(vx)
ridge_score = ridge.score(vx,vy)
coeff_used = np.sum(ridge.coef_ != 0)
ridge.coef_
print("L2 Regression:")
print("R-square score:", r2_score(rp,vy))
print("training score",ridge_score)
print("number of feature used:",coeff_used)
```

```python
from sklearn.linear_model import Lasso 
lasso = Lasso()
lasso
lasso.fit(tx,ty)
lp=lasso.predict(vx)
lasso_score = lasso.score(vx,vy)
coeff_used = np.sum(lasso.coef_ != 0)
lasso.coef_
print("L1 Regression:")
print("R-squared score:",r2_score(lp,vy))
print("training score",lasso_score)
print("number of feature used:",coeff_used)
```
